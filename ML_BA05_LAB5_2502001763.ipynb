```python
#library
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df1 = pd.read_excel('dataset_lab5_2.xlsx')
df2 = pd.read_excel('dataset_lab5_1.xlsx')
```

```python
from sklearn.cluster import KMeans

inertia = []
K = range(1,10)
for i in K:
  kmeans = KMeans(n_clusters = i)
  kmeans.fit(df1)
  y_kmeans = kmeans.predict(df1)
  inertia.append(kmeans.inertia_)

```

```python
plt.plot(K,inertia)
plt.xlabel('K')
plt.ylabel('Inertia')
plt.show()
```

```python
kmeans = KMeans(n_clusters = 4)
kmeans.fit(df1)
y_kmeans = kmeans.predict(df1)


fig, axs = plt.subplots(2, 2)
fig.set_figwidth(15)
fig.set_figheight(6)
sns.scatterplot(ax=axs[0,0], x=df1['A'], y=df1['B'], hue = y_kmeans)
sns.scatterplot(ax=axs[0,1], x=df1['B'], y=df1['C'], hue = y_kmeans)
sns.scatterplot(ax=axs[1,0], x=df1['C'], y=df1['D'], hue = y_kmeans)
sns.scatterplot(ax=axs[1,1], x=df1['D'], y=df1['A'], hue = y_kmeans)
axs[0, 0].set_xlim(-15, 15)
axs[0, 0].set_ylim(-15, 15)
axs[0, 1].set_xlim(-15, 15)
axs[0, 1].set_ylim(-15, 15)
axs[1, 0].set_xlim(-15, 15)
axs[1, 0].set_ylim(-15, 15)
axs[1, 1].set_xlim(-15, 15)
axs[1, 1].set_ylim(-15, 15)
axs[0, 0].set_title('ColA vs ColB')
axs[0, 0].set_xlabel('ColA')
axs[0, 0].set_ylabel('ColB')
axs[0, 1].set_title('ColB vs ColC')
axs[0, 1].set_xlabel('ColB')
axs[0, 1].set_ylabel('ColC')
axs[1, 0].set_title('ColC vs ColD')
axs[1, 0].set_xlabel('ColC')
axs[1, 0].set_ylabel('ColD')
axs[1, 1].set_title('ColD vs ColA')
axs[1, 1].set_xlabel('ColD')
axs[1, 1].set_ylabel('ColA')

plt.show()
```

```python
from sklearn.mixture import GaussianMixture
from sklearn.cluster import DBSCAN

kmeans = KMeans(n_clusters = 5)
kmeans.fit(df2)
y_kmeans = kmeans.predict(df2)

gaussian = GaussianMixture(n_components = 5)
gaussian.fit(df2)
y_gaussian = gaussian.predict(df2)

dbscan = DBSCAN(eps = 1.5).fit(df2)
y_dbscan = dbscan.labels_


fig, axs  = plt.subplots(1,3)
fig.set_figwidth(15)
fig.set_figheight(6)
sns.scatterplot(ax=axs[0], x=df2['A'], y=df2['B'], hue = y_kmeans)
sns.scatterplot(ax=axs[1], x=df2['A'], y=df2['B'], hue = y_gaussian)
sns.scatterplot(ax=axs[2], x=df2['A'], y=df2['B'], hue = y_dbscan)
axs[0].set_title('kmeans')
axs[1].set_title('Gaussian Mixture')
axs[2].set_title('DBSCAN')

plt.show()
```
